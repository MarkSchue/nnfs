```python
import numpy as np

def softmax(x):
    e_x = np.exp(x- np.max(x))
    return e_x / e_x.sum(axis=0)


x = np.linspace(-5,5,100)

y = softmax(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Softmax Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of x values
x = np.linspace(-3, 3, 100)  # Creates 100 evenly spaced points between -3 and 3

# Calculate the corresponding y values using the exponential function
y = np.exp(x)  # np.exp() is the function for e^x

# Create the plot
plt.plot(x, y)

# Add labels and title (optional but good practice)
plt.xlabel("x")
plt.ylabel("e^x")
plt.title("Plot of e^x")

# Add a grid (optional)
plt.grid(True)

# Show the plot
plt.show()
```
